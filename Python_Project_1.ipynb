```python
import pandas as pd
import pyspan as ps
import numpy as np
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_columns', None)

df = pd.read_csv('/content/train.csv')

df.head()
```

```python
df = ps.remove(df, operation= 'columns', columns= 'SSN')

df.head()
```

```python
pd.set_option('display.max_rows', None)

df.isnull().sum()
```

```python
df = ps.handle_nulls(df, columns=['Name', 'Monthly_Inhand_Salary', 'Type_of_Loan', 'Num_of_Delayed_Payment', 'Num_Credit_Inquiries', 'Credit_History_Age', 'Amount_invested_monthly', 'Monthly_Balance'], action='replace', with_val='No')

df.head()
```

```python
df.replace(['No', '-'], np.nan, inplace=True)

df.head()
```

```python
df['Age'].value_counts()
```

```python
df.replace('_', inplace=True)

df.head()
```

```python
df['Age'].value_counts()
```

```python
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')

df.dropna(subset=['Age'], inplace=True)

df['Age'].value_counts()
```

```python
df['Age'] = df['Age'].astype(int)

df['Age'].value_counts()
```

```python
df = df[df['Age'] < 95]

df = df[df['Age'] >= 0]

df['Age'].value_counts()
```

```python
df.head()
```

```python
df.shape
```

```python
df = ps.detect_outliers(df, columns = ['Age'], method = 'iqr', threshold = 3 )
```

```python
df = ps.split_column(df, column='Credit_History_Age', delimiter= ' and ')

df.head()
```

```python
test = df['Credit_Score'].value_counts().reset_index()

test
```

```python
rider_breakdown = df['Credit_Score'].value_counts()

x_axis = rider_breakdown.index
y_axis = rider_breakdown.values

plt.bar(x_axis, y_axis, color = 'red')
plt.title("Credit Remarks")
plt.ylabel("Count")
plt.xlabel("Remarks")
```

```python
plt.pie(y_axis, labels = x_axis, autopct='%1.1f%%')
```
